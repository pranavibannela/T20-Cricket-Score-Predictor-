```python
#cities with minimum 600 balls bowled
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist() 
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
print(df['player_dismissed'].dtype)
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
# Convert '0' to integer directly within .loc
df.loc[df['player_dismissed'] == '0', 'player_dismissed'] = 0

# Convert the entire column to integer (assuming other values are numeric)
df['player_dismissed'] = df['player_dismissed'].astype(int)

# Now you can use cumsum on numeric data
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()

# Calculate remaining wickets
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
df
```

```python
def is_numeric(x):
  try:
    float(x)
    return True
  except ValueError:
    return False

# Filter only numeric values in 'runs' for each group
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
  group_df = groups.get_group(id)
  # Filter numeric values in 'runs' column
  numeric_runs = group_df[group_df['runs'].apply(is_numeric)]['runs']
  # Now you can use rolling.sum on numeric data
  last_five.extend(numeric_runs.rolling(window=30).sum().values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five']]
```

```python
df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error

import xgboost   #xgboost is giving better results
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
import xgboost
xgboost.__version__ 
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
